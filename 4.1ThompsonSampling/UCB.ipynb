```python
#dataset me 
#   ad1 ad2 ad3 .......
#1   1   0   0 --->fidfferent types of ads if clicked or not 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
import random
N=1000
d=10
ads_selected=[]
number_of_reward_1=[0]*d
number_of_reward_0=[0]*d
total_reward=0
for n in range(0,N):
   ad=0
   max_random=0
   for i in range(0,d):
      random_beta=random.betavariate(number_of_reward_1[i]+1,number_of_reward_0[i]+1)
      if(random_beta>max_random):
         max_random=random_beta
         ad=i #selcting the best reslt for working for user 1 ad 1 
   ads_selected.append(ad) 
   reward=dataset.values[n,i]
   if(reward==1):
      number_of_reward_1[ad]+=1
   else:
      number_of_reward_0[ad]+=1
   total_reward+=reward   


       
 

      
```

```python
plt.hist(ads_selected)
plt.title('Histogram f ads')
plt.xlabel('ads')
plt.ylabel('nUMBER OD TIMES CLCIKED')
plt.show()
```
