```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('50_Startups.csv')
#print(dataset)
x=dataset.iloc[:,:-1].values
y=dataset.iloc[:,-1].values
#print(x)
```

```python
#encding fs ke baad spliting hti hai 
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct=ColumnTransformer(transformers=[('encoder',OneHotEncoder(),[3])],remainder='passthrough')
X=np.array(ct.fit_transform(x))
#print(X)
```

```python

from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
#import the right class to implement l regression model
from sklearn.linear_model import LinearRegression
regressor=LinearRegression() #modle buuilt now apply it 
regressor.fit( X_train,Y_train) #fit model to ut model


```

```python
#predict the results
y_pred=regressor.predict(X_test)  #train kr diya by test 
#y_test me actul hai 
#y_pred me test set ke data hai wo model ne trin set obserbe krke kiya hi 

#print(y_pred)
#print(Y_test)
```

```python
#visualise results
plt.scatter(X_train,Y_train,color='red')
#now the rgrssor line
plt.plot(X_train,regressor.predict(X_train))
plt.title('Salary vs Exp')
#similar for test set 
```
