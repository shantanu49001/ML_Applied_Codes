```python
#dataset me 
#   ad1 ad2 ad3 .......
#1   1   0   0 --->fidfferent types of ads if clicked or not 
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Ads_CTR_Optimisation.csv')
```

```python
import math
N=1000
d=10
ads_selected=[]
num_of_elections=[0]*d #10 adds  hr round pe agar yeadd clicked hoga to value incr ho jayegi
sum_of_rewards=[0]*d
total_reward=0
for n in range(0,N):
    ad=0
    max_upper_bound=0
    for i in range(0,d):
        if (num_of_elections[i]>0):
            average_reward=sum_of_rewards[i]/num_of_elections[i]
            delata_i=math.sqrt(3/2*math.log(n+1))/num_of_elections[i]
            upper_bound=average_reward+delata_i  #upper bound of ith add in ith round
        else: #ek bhi baar click nhi bua 
            upper_bound=1e400
        if(upper_bound>max_upper_bound):  #update for the ith add 
            max_upper_bound=upper_bound
            ad=i     #select add with max upper bound 

    ads_selected.append(ad)
    num_of_elections[ad]=num_of_elections[ad]+1
    reward=dataset.values[n,ad]
    sum_of_rewards[ad]=reward+sum_of_rewards[ad]  #user ad 
    total_reward+=reward


      
```

```python
plt.hist(ads_selected)
plt.title('Histogram f ads')
plt.xlabel('ads')
plt.ylabel('nUMBER OD TIMES CLCIKED')
plt.show()
```
