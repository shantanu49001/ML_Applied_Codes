```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
X=dataset.iloc[:,[3,4]].values  #all values and 3rd fourth col
#  Y=dataset.iloc[:,-1].values
```

```python
#elbow mwthod to draw clusters
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kMeans=KMeans(n_clusters=i,init= 'k-means++',random_state=42)       #10 clusters
    kMeans.fit(X)
    wcss.append(kMeans.inertia_)
plt.plot(range(1,11),wcss)
    
```

```python
kMeans=KMeans(n_clusters=5,init='k-means++',random_state=42)
y_means=kMeans.fit_predict(X)
print(y_means)
```

```python
plt.scatter(X[y_means==0,0],X[y_means==0,1],s=100,c='red',label='Cluster1')
plt.scatter(X[y_means==1,0],X[y_means==1,1],s=100,c='blue',label='Cluster1')
plt.scatter(X[y_means==2,0],X[y_means==2,1],s=100,c='green',label='Cluster1')

plt.scatter(kMeans.cluster_centers_[:,0],kMeans.cluster_centers_[:,1],s=300,c='yellow')
```
