```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Position_Salaries.csv')   #i have to redict salary based on 
#print(dataset)   #first to last
x=dataset.iloc[:,1:-1].values  #names wala col hta do
y=dataset.iloc[:,-1].values
```

```python
#import the right class to implement l regression model
from sklearn.linear_model import LinearRegression
lin_reg=LinearRegression() #modle buuilt now apply it 
lin_reg.fit( x,y) #fit model to ut model



```

```python
#building the polyn regression m
from sklearn.preprocessing import PolynomialFeatures
ploy_reg=PolynomialFeatures(degree=2) #2nd max  poee tak
X_poly=ploy_reg.fit_transform(x)
lin_reg2=LinearRegression()
lin_reg2.fit(X_poly,y)
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg.predict(x),color='blue')
```

```python
plt.scatter(x,y,color='red')
plt.plot(x,lin_reg2.predict(ploy_reg.fit_transform(x)),color='blue')
```
