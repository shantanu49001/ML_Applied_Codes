```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset=pd.read_csv('Mall_Customers.csv')
X=dataset.iloc[:,[3,4]].values  #all values and 3rd fourth col
#  Y=dataset.iloc[:,-1].values
```

```python
import scipy.cluster.hierarchy as sch
dendogram=sch.dendrogram(sch.linkage(X,method='ward'))
plt.title('Dendogram')
plt.xlabel('Customers')  #x axiz is thr wor not the two pount 
plt.ylabel('Euclid Distance')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering
hc=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
y_hc=hc.fit_predict(X)
```

```python
plt.scatter(X[y_hc==0,0],X[y_hc==0,1],s=100,c='red',label='Cluster1')
plt.scatter(X[y_hc==1,0],X[y_hc==1,1],s=100,c='blue',label='Cluster1')
plt.scatter(X[y_hc==2,0],X[y_hc==2,1],s=100,c='green',label='Cluster1')

#plt.scatter(kMeans.cluster_centers_[:,0],kMeans.cluster_centers_[:,1],s=300,c='yellow')
```
